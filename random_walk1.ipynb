```python
import numpy as np

N = 1000000

r = np.random.random(N+3)

#za br [0,1] nezavisnost:  <r_i>= o.5  pa <r_i r_i+3> =0.5 *0.5=0.25

products = r[:N] * r[3:N+3]

cumsum = np.cumsum(products)  

i = np.arange(1, N+1)

temp = np.abs(cumsum/i - 0.25) * np.sqrt(i)  #odstupanje C(3) od idealne vr 0.25

data = np.column_stack((i, temp))

np.savetxt("z1_rez.txt", data[::10], fmt="%7d %13.8e")
```

```python
import matplotlib.pyplot as plt

data = np.loadtxt("z1_rez.txt")

plt.plot(data[:,0], data[:,1])
plt.xlabel("N")
plt.ylabel("sqrt(N)*|C(3)-0.25|")
plt.show()
```
